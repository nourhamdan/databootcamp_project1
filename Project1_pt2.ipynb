```python
import pandas as pd
import requests
from IPython.display import clear_output
import matplotlib.pyplot as plt
```

```python
#read USDA crop yield CSV data file
df = pd.read_csv('crop_yields.csv')
```

```python
df['Year']=df['Year'].astype('str')
```

```python
#clean up the data by removing unnecessary columns and merging COUNTY_STATE
df['COUNTY_STATE']=df['County']+', '+df['State']
df=df.drop(columns=['Program','Period','Week Ending','Geo Level','State ANSI', 'Ag District', 'Ag District Code','County ANSI','Zip Code','Region','watershed_code','Data Item','Domain','Domain Category','CV (%)'])
df.head()
```

```python
#create data frame just for corn yield data
df_corn = df.loc[df['Commodity']=='CORN']
df_corn = df_corn.rename(columns={'Value':'Corn Yield'})
df_corn=df_corn.drop(columns='Commodity')
df_corn.head()
```

```python
#create data frame just for soybeans yield data
df_soy = df.loc[df['Commodity']=='SOYBEANS']
df_soy = df_soy.rename(columns={'Value':'Soybean Yield'})
df_soy=df_soy.drop(columns='Commodity')
df_soy.head()
```

```python
#create data frame just for wheat yield data
df_wheat = df.loc[df['Commodity']=='WHEAT']
df_wheat = df_wheat.rename(columns={'Value':'Wheat Yield'})
df_wheat=df_wheat.drop(columns='Commodity')
df_wheat.head()
```

```python
#merge corn and wheat yield dataframes
df2=pd.merge(df_corn, df_wheat, on=['COUNTY_STATE','Year','County','State'], how="outer")
df2.head()
```

```python
#merge soybeans to corn and wheat yield dataframe
df3=pd.merge(df2,df_soy, on = ['COUNTY_STATE','Year','County','State'], how="outer")
df3=df3.drop(columns=['Watershed_x','Watershed_y','Watershed'])
df3.head()
```

```python
#make 5 dataframes including all 3 crop yields, one dataframe per state
df_mo = df3.loc[df3['State']=='MISSOURI']
df_il=df3.loc[df3['State']=='ILLINOIS']
df_tx=df3.loc[df3['State']=='TEXAS']
df_ia=df3.loc[df3['State']=='IOWA']
df_ks=df3.loc[df3['State']=='KANSAS']
```

```python
df_mo.to_csv("mo_yields.csv")
df_il.to_csv("il_yields.csv")
df_tx.to_csv("tx_yields.csv")
df_ia.to_csv("ia_yields.csv")
df_ks.to_csv("ks_yields.csv")
```
